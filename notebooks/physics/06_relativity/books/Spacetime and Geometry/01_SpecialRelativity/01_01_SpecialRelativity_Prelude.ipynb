## Newtonian gravity recap

Before relativity, gravity is modeled as a force acting in space over time. The
next two cells show the familiar constant-acceleration example and the Poisson
equation for the gravitational potential.


```python
from sympy.physics.mechanics import dynamicsymbols
from sympy.vector import CoordSys3D

N = CoordSys3D('N')
t = sp.symbols('t')

x, y, z = dynamicsymbols('x y z')
m, F = dynamicsymbols('m F')

x, y = 0, 0

z = 0.5 * -9.8 * t ** 2

r_vec = x*N.i + y*N.j + z*N.k

a = r_vec.diff(t, 2)

sp.Eq(F, sp.UnevaluatedExpr(m*a), evaluate=False)
```

```python
from sympy.physics.units import gravitational_constant as G

x, y, z = sp.symbols('x y z')
phi = sp.Function('phi')(x, y, z)
rho = sp.Function('rho')(x, y, z)

pi = sp.pi
poisson = (
    sp.diff(phi, x, 2)
  + sp.diff(phi, y, 2)
  + sp.diff(phi, z, 2)
  - 4 * pi * G * rho
)

sp.Eq(poisson, 0)
```
