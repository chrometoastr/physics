### Calculus of Variations

## Core objects
- Generalized coordinates: $q^i(t)$
- Lagrangian: $L(q, \dot q, t)$
- Action: $S[q] = \int_{t_1}^{t_2} L(q,\dot q,t)\,dt$

## Principle of stationary action
$\delta S = 0$ (with fixed endpoints $\delta q^i(t_1)=\delta q^i(t_2)=0)$
⇒ Euler–Lagrange:
$\frac{d}{dt}\left(\frac{\partial L}{\partial \dot q^i}\right) - \frac{\partial L}{\partial q^i} = 0$

```python
# Suggested imports (keep it consistent across notebooks)
import sympy as sp
sp.init_printing()

import numpy as np
```

## Example: Harmonic oscillator from the action
$$
L = \frac{1}{2}m\dot x^2 - \frac{1}{2}k x^2
E–L ⇒ m\ddot x + kx = 0
$$

```python
t = sp.symbols('t', real=True)
m, k = sp.symbols('m k', positive=True, real=True)
x = sp.Function('x')(t)

xd = sp.diff(x, t)
L = sp.Rational(1,2)*m*xd**2 - sp.Rational(1,2)*k*x**2

EL = sp.diff(sp.diff(L, xd), t) - sp.diff(L, x)
sp.simplify(EL)
```
