## Introduction

Classical Mechanics is a set of rules about what the Laws of Motion look like. What are the questions about the laws of motion?

- What are the specific laws for a particular system?  Like, what are the laws of motion for a planet moving in the field of a heavy mass?  Those laws may be different that an electrically charged particle moving in a magnetic field.
- What are the rules for the allowable laws?  What is the grand framework for all the various different specific laws are framed

The first question gives us examples of the answer to the second question. What are the allowable laws of physics?

Let's start with a system consisting of a coin.

## Law of Motion
What are the laws of motion for a coin?

Answer: it stays the same. 

$$
H -> H,
T -> T
$$

or
$$
H -> T,
T -> H
$$
                     

```python
from graphviz import Digraph

dot = Digraph()

dot.node("H", "Heads")
dot.node("T", "Tails")

dot.edge("H", "T")
dot.edge("T", "H")

dot
```

HTHTHTHTHTHTHTHTHTHTH...
THTHTHTHTHTHTHTHTHTHT...

Create an equation of motion where the coin never flips. Use $\sigma$ because it is typically used to denote either $1$ or $-1$.:
$$\sigma(t+1) = \sigma(t)$$
In a stroboscopic world, time is an integer (like a strobe light).
What about when the coin does flip after every flash of the strobe light?  The equation becomes:
$$\sigma(t+1) = -\sigma(t)$$

Note the that system is completely predictive. There is no ambiguity about what the state of the coin would be arbitrarily into the future.  This is a closed system. If I come in and interfere and switch the coin side, then it is not a closed system.

What about a Die?

A die has six states: $(1, 2, 3, 4, 5, 6)$

An initial condition is a choice of one of the six states.

What about the laws of motion?

A simple law is nothing happens. A more interesting law is that you cycle through all of the states:


```python
dot = Digraph()

dot.node("1", "1")
dot.node("2", "2")
dot.node("3", "3")
dot.node("4", "4")
dot.node("5", "5")
dot.node("6", "6")

dot.edge("1", "2")
dot.edge("2", "3")
dot.edge("3", "4")
dot.edge("4", "5")
dot.edge("5", "6")
dot.edge("6", "1")

dot
```

What would the equation of motion be?

What would the history of the world be?  Given a starting point, say 4, the history of the world or the system, would be $4,5,6,1,2,3,...$, endlessly cycling.

That would be the theory of this particular die.

- Another law of motion would be for the die to cycle through in the opposite direction.
- Yet another law, is 1,2,5,3,4,6,1,2,...  It is the same as the other laws, it's just a renaming of the states or cycle.  It's logically equivalent to the others.

## Conservation Law

Here'a law that is not equivalent to the others:

```python
dot = Digraph()

dot.node("1", "1")
dot.node("2", "2")
dot.node("3", "3")
dot.node("4", "4")
dot.node("5", "5")
dot.node("6", "6")

dot.edge("1", "2")
dot.edge("2", "3")
dot.edge("3", "1")
dot.edge("4", "5")
dot.edge("5", "6")
dot.edge("6", "4")

dot
```

There two cycles now (given a name "cycles").  This particular law of physics means that once the system is one cycle, it cannot move into the other cycle.
There is a name for this behaviour `Conserved Quantity`.  

A `Conserved Quantity` is something non-trivial you can label a system **that does not vary with time**. 

If you label the first graph with the value "1", and the second graph with value "2", then a `Conserved Quantity` of the system could be if you start with the conserved quantity labelled "1", then you stay with "1" forever.  If you start with the conserved quantity "2", then you stay in conserved quantity "2" forever.

This doesn't change with time.  This is a `Conservation Law`.

The state of the system or the configuration consists of all the things you need to know to predict the future state of the system.

So far, we the concepts of:
- Configurations (states)
- Isolated Systems
- Conservation laws

There's nothing saying we cannot have an infinite number of states.  Consider a particle on the integer number line.  This single particle can have an infinite number of states.  And we can invent laws of motion.  One law of motion could be if you are at position $n$, then jump to $n+1$. It does not cycle in this case.  There are no interesting Conservation Laws in this case. There is no conserved quantity because the particle traverses through each state.  

For a second law of motion, if the particle is at position $n$, the jump to $n+2$.  In this case, there is a `Conserved Quantity`. In this case, oddness or evenness.  Once you start out with an even state, you stay on even numbers, and if you start on an odd state then you stay on odd numbers.

## Allowable Rules
These are rules that are allowed by the system. For the die case, the rules are adhoc, but there are allowable rules in nature.

This is a kind of law that we would not allow in `Classical Physics`:

```python
from graphviz import Digraph

dot = Digraph()

dot.node("H", "Heads")
dot.node("T", "Tails")
dot.node("E", "Edge")

dot.edge("H", "T")
dot.edge("T", "E")
dot.edge("E", "T")

dot
```

What is odd about this law?

- It is completely predictive into the future, but not into the past.
- You can predict the future but you cannot retrodict the past.
- The system is not reversible.

If you started from Tails, did you come from "Edge" or from "Heads"?

The "Heads" and "Edge" configuration both run into the "Tails" configuration.

By reversing the directions of the edges, the system becomes `non-reversible`:



```python
from graphviz import Digraph

dot = Digraph()

dot.node("H", "Heads")
dot.node("T", "Tails")
dot.node("E", "Edge")

dot.edge("T", "H")
dot.edge("T", "E")
dot.edge("E", "T")

dot
```

Now you have an unpredictive situation.  You have a unpredictive situation.

This is `Not Reversible` This is also not allowed in classical physics.

## Limits on Predictability
There are limits to how well you can specify the initial conditions of a system because in the real world, initial conditions, like location, are specified with real numbers ($x \in \mathbb{R}$). How fast the predictive nature of a system because of this imperfect knowledge may or may not allow you to predict the future in a useful way.

If you knew how imperfect your knowledge of the initial conditions are, quantitatively, then you may be able to predict how far into the future (or past) you can predict accurately. This is a true classical mechanics system. So if you wanted to predict several years into the future, you would to know how precise you need to be to achieve this accuracy.

## How to determine if a Law of Motion is Valid
The laws of classical mechanics are not only deterministic into the future but deterministic into the past.  This makes the system reversible.

How do you look at these configuration diagrams (law of motion) and determine if it is valid or not? If you can't tell where you came from (two arrows coming into a configuration), or you can't tell where you are going (two arrows leaving a configuration), then the laws of motion are not valid.
